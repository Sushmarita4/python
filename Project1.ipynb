```python
df_sorted = df.sort_values('Nitrogen', ascending=False).reset_index()

# Create the bar plot
plt.figure(figsize=(6,6), dpi=50)
sns.barplot(x='Nitrogen', y='Crop', data=df_sorted)
plt.xlabel('Average Nitrogen')
plt.ylabel('Crop')
plt.title('Average Nitrogen by Crop')
plt.show()
```
